### Importing Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Loading Dataset

```python
df = pd.read_csv("updated_ullu_metadata (1).csv")
```

```python
df.head()
```

### Exploratory Data Analysis

```python
df.info()
```

```python
df.isna().sum()
```

### Data Visualization

```python
plt.figure(figsize=(15,6))
plt.hist(df['Language'],bins=20,color="red",edgecolor="black")
plt.title('Distribution of Languages', fontsize=16)
plt.xlabel("Languages")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['Language'].value_counts().plot.pie(autopct='%1.1f%%', wedgeprops=dict(width=0.3, edgecolor='black'))
plt.title('Distribution of Languages', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['Year'],bins=10,color="orange",edgecolor="black")
plt.title('Distribution of Year', fontsize=16)
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['Year'].value_counts().plot.pie(autopct='%1.1f%%', wedgeprops=dict(width=0.3, edgecolor='black'))
plt.title('Distribution of Year', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df["Search Category"],bins=5,color="green",edgecolor="black")
plt.title("Distribution of Search Category")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df["Freely Available"].value_counts().plot(kind="bar",color="yellow")
plt.title("Distribution of Freely Available")
plt.ylabel("Frequency")
plt.show()
```

## Exploring year 2021

```python
year_2021_language = np.array([])
for year,language in zip(df['Year'],df["Language"]):
    if year == 2021:
        year_2021_language=np.append(year_2021_language,language)
        
plt.figure(figsize=(15,6))
plt.hist(year_2021_language,bins=20,color="blue",edgecolor="black")
plt.title("Distribution of language in 2021")
plt.ylabel("Frequency")
plt.show()
```

```python
year_2021_sf = np.array([])
for year,language,sf in zip(df['Year'],df["Language"],df["Search Category"]):
    if year == 2021 and language == "Hindi":
        year_2021_sf=np.append(year_2021_sf,sf)
        
plt.figure(figsize=(15,6))
plt.hist(year_2021_sf,bins=20,color="indigo",edgecolor="black")
plt.title("Distribution of search category with hindi language in 2021")
plt.ylabel("Frequency")
plt.show()
```
